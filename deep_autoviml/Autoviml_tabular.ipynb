```python
project_name="deep_autoviml"
model_options = {'cat_feat_cross_flag':"",
                 'max_trials': 5, "tuner": "storm"}
keras_options={"patience":1, 'class_weight': True, 'early_stopping': True, 
                 'lr_scheduler': '', "optimizer": 'RMS'}
keras_model_type = "auto"
# file_path = '/content/Train.csv'
file_path='/content/flat-prices.csv'
```

```python
df = pd.read_csv(file_path)
df=df[:10000]
df.head()
```

```python
model, cat_vocab_dict = deepauto.fit(df, "resale_price", keras_model_type, project_name=project_name, keras_options=keras_options,  
		model_options=model_options, save_model_flag=False, use_my_model='', verbose=1)

# model, cat_vocab_dict = deepauto.fit(df, target, keras_model_type,project_name=project_name, keras_options=keras_options,
                                    #  model_options=model_options, save_model_flag=False, use_my_model='', verbose=1)
```

```python
predictions = predict_text(file_path, model, cat_vocab_dict,keras_model_type=keras_model_type)
```
