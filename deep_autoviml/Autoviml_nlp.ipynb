```python
df = pd.read_csv('/content/Train.csv')
df=df[['text','label']]
df.head()
```

```python
target = 'label'
project_name = 'nlp'
keras_model_type =  "USE"
keras_options = {"early_stopping": True, 'epochs': 10, 'steps_per_epoch': 15, 'class_weight': True}
model_options = {'tuner':"storm", "max_trials":10}
```

```python
from deep_autoviml.modeling.predict_model import predict_text
test_file_path = '/content/Train.csv'
predictions = predict_text(test_file_path, model, dicti,keras_model_type=keras_model_type)
```
